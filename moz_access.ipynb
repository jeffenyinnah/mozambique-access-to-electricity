## Converting datasets from wide to long dataset type

```python
## naming conventions
## urban = Access to electricity, urban (% of urban population)
## rural = Access to electricity, rural (% of rural population)
## population = Access to electricity (% of population)

# to convert from wide to long, we will use melt() function

urban = pd.melt(urban, id_vars =['Country Name', 'Country Code', 'Indicator Name','Indicator Code'], var_name='Date',
             value_name='urban')

rural = pd.melt(rural, id_vars =['Country Name', 'Country Code', 'Indicator Name','Indicator Code'], var_name='Date',
             value_name='rural')
population = pd.melt(population, id_vars =['Country Name', 'Country Code', 'Indicator Name','Indicator Code'], var_name='Date',
             value_name='population')
```

```python
urban.head()
```

```python
rural.head()
```

```python
population.head()
```

## Merge all the datasets together to get a combined dataset for further analysis

```python
df = pd.merge(population, urban,  how='left', left_on=['Country Name','Country Code', 'Date'], 
         right_on = ['Country Name','Country Code', 'Date'])
```

```python
df = pd.merge(df, rural, how = 'left', left_on=['Country Name','Country Code', 'Date'], 
              right_on=['Country Name','Country Code', 'Date'])
df.head()
```

## Selecting the colunms i need for my analysis
### The focus of this analysis is the Country Mozambique. We will select all records of Mozambique.

```python
df = df[df['Country Name'] == "Mozambique"]
```

```python
data = df[['Country Name','Country Code', 'Date', 'population', 'urban', 'rural']]
data.head()
```

### Now we have the columns of interest, lets carry out some analysis

```python
# First, lets change the 'Date' column to actual Date format

print(data.dtypes)
```

```python
# from string to datetime format

data['Date'] = pd.to_datetime(df['Date'])
data.info()
```

## Visualizing Missing Data

### Visualizing Missing Data using Seaborn heatmap()

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("visualizing_missing_data_with_heatmap_Seaborn_Python.png", dpi=100)
```

### Visualizing Missing Data using Seaborn displot()

```python
plt.figure(figsize=(10,6))
sns.displot(
    data= data.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25
)
plt.savefig("visualizing_missing_data_with_barplot_Seaborn_distplot.png", dpi=100)
```

### From the above graphs we can see there majority of missing values. Meanwhile, dataset contains data from 1960 and we are goning to leave the dataset as it is because we want to do a timeseries from 1960 to recent date. 
- True shows the level of missing values.
- False shows non missing values.
